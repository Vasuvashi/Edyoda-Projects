# Design a Food Ordering app for a restaurant

```python
import re
import random
import json
class Restaurant:
    def __init__(self, name):
        self.Restaurant_name = name
        self.food_id = random.randint(1,1000)
        self.id = random.randint(1,100)
        self.ordered_item = []
        try:
            with open("User_Details.txt","r") as file:
                user_data = json.load(file)
                self.all_user_details = user_data
        except Exception as ex:
            self.all_user_details = {}
        
        self.user_details = {}
            
    def valid_email(self,email):
        pattern = r"^\w*@\w*[.]\w*$"
        valid = re.findall(pattern, email)
        if valid:
            print("'Valid Email-id'")
        else:
            print("Invalid Email-id.\nPlease enter a Valid Email-id")
            email = self.valid_email(input("Email:"))
        return email
    
    def valid_contact(self,phone_number):
        pattern = r"^[6-9]\d{9}"
        number = re.findall(pattern ,phone_number)
        if number:
            print("Validated!")
        else:
            print("Invalid Phone Number.\n Please enter a valid Phone Number")
        return number
    
    def valid_pass(self, password):
        pattern = "^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)[a-zA-Z\d\w\W]{8,}$"
        valid = re.findall(pattern, password)
        print(valid)
        if valid:
            print("Valid password")
        else:
            print("Invalid password")
        return password
```

```python
class FoodApp_admin(Restaurant):
    def __init__(self, user_id, password, name):
        super().__init__(name)
        self.user_id = user_id
        self.password = password
        try:
            with open("food_items.txt","r") as file:
                self.data = json.load(file)
                self.all_foodItems = self.data
        except Exception as ex:
            self.all_foodItems = {}

        self.foodItems = {}
        if self.user_id == "admin" and self.password == "admin@123":
            print(f"---------------------Hi Admin, Welcome to the {self.Restaurant_name} Restaurant---------------------")
        else:
            print("-----Incorrect Login details-----")

    def __repr__(self):
        if self.user_id == "admin" and self.password == "admin@123":
            return "True"
        else:
            return "False"

    def add_items(self, Food_name, Qty, Price, Discount, Stock):
        self.foodItems["Food_name"] = Food_name
        self.foodItems["Qty"] = Qty
        self.foodItems["Price"] = Price
        self.foodItems["Discount"] = Discount
        self.foodItems["Stock"] = Stock
        
        while True:                
            if self.food_id not in self.all_foodItems.keys():
                self.all_foodItems[self.food_id] = self.foodItems
                with open("food_items.txt","w") as file:
                    json.dump(self.all_foodItems, file)
                print("Food items added Successfully!!")
                break
            else:
                self.food_id += 1

    def update_food_items(self, food_id, parameter, new_value):
        if food_id in self.all_foodItems.keys():
            self.food_Itemdetails = self.all_foodItems[food_id]
            if parameter in self.food_Itemdetails.keys():
                self.food_Itemdetails[parameter] = new_value
                self.all_foodItems[food_id] = self.food_Itemdetails
                with open("food_items.txt","w") as file:
                    json.dump(self.all_foodItems, file)
                print("Items Updated Successfully!!")

            else:
                print("This is not a valid Parameter.")

        else:
            print("Food Id not Found!!!")

    def display_food_items(self):
        print("------------------Food Items------------------\n")
        for k,v in self.all_foodItems.items():
            print(k, "------->", v)

    def delete_foodItem(self, food_id):
        try:
            if food_id in self.all_foodItems.keys():
                del self.all_foodItems[food_id]
                with open("food_items.txt","w") as file:
                    json.dump(self.all_foodItems, file)
                print("Food Item removed from the app Successfully!!")
            else:
                print("Food Id Doesn't Exist!")
        except Exception as ex:
            print("No Food Item present to be removed from the list.")
```

```python
class FoodApp_User(Restaurant):
    def __init__(self,name):
        super().__init__(name)
        print(f"Hi There!! Welcomee to {self.Restaurant_name} Please register Yourself to enjoy the facilities.\n if You are already our customer. Please login yourself")
        
    def user_register(self):
        try:
            self.user_details["Full_name"] = input("Enter your Full Name: ")
            self.user_details["Email-ID"] = self.valid_email(input("Enter your Email-ID: "))
            self.user_details["Password"] = self.valid_pass(input('''Enter a password\nPassword should contain 1 Ucase, 1 Lcase, 1 special character, 1 digit and limit should be minimum 8 characters\n'''))
            self.user_details["Phone_Number"] = self.valid_contact(input("Enter your 10 digit Phone number: "))
            self.user_details["Address"] = input("Enter your Address with ZIP code: ")
            
            while True:
                if self.id not in self.all_user_details.keys():
                    self.all_user_details[self.id] = self.user_details
                    with open("User_Details.txt","w") as file:
                        json.dump(self.all_user_details, file)
                    print("\n!! Your Account has been created Successfully !!\n")
                    print(f"---------------Welcome to the {self.Restaurant_name} Restaurant---------------\n")
                    print("User Details: \n")
                    print ("User ID",self.id)
                    for k,v in self.user_details.items():
                        print(k, " : " ,v)
                    
                    break
                else:
                    self.id +=1
        
        except Exception as ex:
            print("\n!!Please provide valid inputs!!\n")
 
    def user_login(self,id):
        try:
            self.id = id
            with open("User_Details.txt","r") as file:
                data = json.load(file)
                self.all_user_details = data
            print(f"---------------Welcome to the {self.Restaurant_name} Restaurant---------------\n")
            email = input("Enter your Email-ID: ")
            pas = input("Enter your Password: ")
            if len(self.all_user_details.values())!=0:
                if email==self.all_user_details[self.id]["Email-ID"] and pas==self.all_user_details[self.id]["Password"] :
                    print("\n!! Login Successfully !!")
                    while True:
                        print("Select any one from the below options\n")
                        print("1. Place New Order\n2. Order History\n3. Update Profile\n4. Delete Profile\n5. Exit\n")
                        x = input()
                        if x == "1":
                            self.Place_Order()
                        elif x == "2":
                            self.Order_History()
                        elif x == "3":
                            self.Update_Profile()
                        elif x == "4":
                            self.delete_profile(input("User_Id: "))
                        elif x == "5":
                            break
                        else:
                            print("Invalid selection")
                else:
                    print("\n!! Incorrect email or password !!\n")
            else:
                print("\n! No such Account exist with this email and ID !\n!! Please register yourself !!\n")
        except Exception as ex:
            print("\n!! Please Provide Valid Inputs !!")

    def Place_Order(self):
        try:
            with open("food_items.txt","r") as file:
                data = json.load(file)
                self.all_foodItems = data
            if len(self.all_foodItems) != 0:
                print("\nList of Available Items: ")
                menu = []
                for i in self.all_foodItems:
                    menu.append([self.all_foodItems[i]["Food_name"], self.all_foodItems[i]["Qty"], self.all_foodItems[i]["Price"]])
                for j in range(len(menu)):
                    print(j+1, ":", menu[j])
                while True:
                    print("\nEnter 1 to place the Order\n 2 to Exit")
                    x = input()                  
                    if x == "1":
                        print("Enter the Food ID You want to Order separated by comma")
                        y = input().split(",")
                        for i in y:
                            z = int(i)
                            if z <= len(menu):
                                self.ordered_item.append(menu[z-1])
                            else:
                                print('''Sorry!! We don't have this food currently. We appreciate to know your feedback 
                                about this food item and add this item in our list!! ''')
                        print("List of Food Items you've selected!!\n")
                        for j in self.ordered_item:
                            print(j)
                        print("\nSelect 1 to Place this order\n 2 to cancel this order.")
                        order = input()
                        if order == "1":
                            print("!! Order Place Successfully !!")
                        elif order == "2":
                            break
                        else:
                            print("!! Invalid Input !!")
                        break
                    elif x == "2":
                        break
                    else:
                        print("\n!! Invalid selection !!\n")
                    break
            else:
                print("\n!! Sorry! No Food Items are available Now !! Please try Another time.\n")
        except Exception as ex:
            print("!! Please provide valid inputs !!")
    
    def Order_History(self):
        print("\nList of Previous ordered : \n")
        for i in self.ordered_item:
            print(i)
                 
    def Update_Profile(self):
        try:
            while True:
                print("Select Below Options to Update Your Profile Details\n")
                print("1. Name\n2. Phone number\n3. Email-ID\n4. Password\n5. Address\n6. Exit\n")
                a = input()
                if a == "1":
                    self.user_details["Full_Name"]=input("Enter your updated full name : ")
                    print("\n!! Detail Updated Successfully !!\n")
                elif a == "2":
                    self.user_details["Phone_Number"]=int(input("Enter your updated 10 digit phone number : "))
                    print("\n!! Detail Updated Successfully !!\n")      
                elif a == "3":
                    self.user_details["Email-ID"]=input("Enter your updated email id : ")
                    print("\n!! Detail Updated Successfully !!")
                elif a == "4":
                    self.user_details["Password"]=input("Enter your updated password : ")
                    print("\n!! Detail Updated Successfully !!\n")
                elif a == "5":
                    self.user_details["Address"]=input("Enter your updated address with area PIN code ")
                    print("\n!! Detail Updated Successfully !!\n")
                elif a == "6":
                    break
                else:
                    print("\n!! Invalid Number Entered !!\n")
                break
                
            if a in ["1","2","3",'4',"5"]:
                for i in self.user_details:
                    print(i, ":",self.user_details[i])
            else:
                print("\nPlease Enter correct Input\n")      
        except Exception as ex:
            print("\n!! Please Provide Valid Inputs !!\n ")
            
    def delete_profile(self,id):
        if self.id in self.all_user_details.keys():
            del self.all_user_details[self.id]
            with open("User_Details.txt","w") as file:
                json.dump(self.all_user_details, file)
            print("Profile Removed Successfully!!")
        else:
            print("ID doesn't Exist!!")
```
